## Display the names of all the employees in Boston 

Hint: join the employees and offices tables.

```python
# Your code here
query = """
SELECT emp.firstName, emp.lastName, off.city FROM employees AS emp 
JOIN offices AS off USING(officeCode)
WHERE off.city = 'Boston'
"""
cur.execute(query).fetchall()
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
# Your code here
query = """
SELECT emp.firstName, emp.lastName, off.city FROM 
offices AS off 
LEFT JOIN employees AS emp USING(officeCode)
WHERE emp.employeeNumber IS null;
"""
cur.execute(query).fetchall()
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here
import pandas as pd
query = """
SELECT e.employeeNumber, e.firstName, e.lastName, 
sum(od.quantityOrdered) AS product_count, p.productName 
FROM products AS p 
JOIN orderdetails AS od USING(productCode)
JOIN orders as o USING(orderNumber)
JOIN customers as c USING(customerNumber)
JOIN employees as e ON e.employeeNumber = c.salesRepEmployeeNumber
GROUP BY e.employeeNumber, p.productCode;
"""
cur.execute(query)
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

## Level Up: Display the Number of Products each employee has sold

```python
# Your code here
import pandas as pd
query = """
SELECT e.employeeNumber, e.firstName, e.lastName, 
count(od.quantityOrdered) AS sale_count 
FROM products AS p 
JOIN orderdetails AS od USING(productCode)
JOIN orders as o USING(orderNumber)
JOIN customers as c USING(customerNumber)
JOIN employees as e ON e.employeeNumber = c.salesRepEmployeeNumber
GROUP BY e.employeeNumber;
"""
cur.execute(query)
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
# Your code here
import pandas as pd
query = """
SELECT e.employeeNumber, e.firstName, e.lastName, 
sum((od.quantityOrdered)*(od.priceEach)) AS sales 
FROM products AS p 
JOIN orderdetails AS od USING(productCode)
JOIN orders as o USING(orderNumber)
JOIN customers as c USING(customerNumber)
JOIN employees as e ON e.employeeNumber = c.salesRepEmployeeNumber
GROUP BY e.employeeNumber;
"""
cur.execute(query)
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```
